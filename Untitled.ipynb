```python
from sklearn.datasets import make_blobs
```

```python
x, y = make_blobs(n_samples=100, n_features=2, random_state=42)
```

```python
k = 3
index = np.random.randint(0, 99, 3)
print(index)
```

```python
c1_cen = x[index[0]]
c2_cen = x[index[1]]
c3_cen = x[index[2]]
print(c1_cen, c2_cen, c3_cen)
```

```python
def euclidean_distance(a, b):
    return np.sqrt( ((a[0] - b[0])**2) + ((a[1] - b[1])**2)  )
```

```python
c1 = []
c2 = []
c3 = []
for i in x:
    d1 = euclidean_distance(c1_cen, i)
    d2 = euclidean_distance(c2_cen, i)
    d3 = euclidean_distance(c3_cen, i)
    m = min(d1, d2, d3)
    if m == d1:
        c1.append(i)
    elif m == d2:
        c2.append(i)
    else:
        c3.append(i)
```
