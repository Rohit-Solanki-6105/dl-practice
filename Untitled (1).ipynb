```python
from tensorflow.keras.utils import image_dataset_from_directory
```

```python
dr = "C:/Users/rcc/Documents/Vehicles"
import os

path = os.path.join(dr)
seed=123

# preprocessing , removing corrupt files
corrupts = 0
import tensorflow as tf
for root, dirs, files in os.walk(dr):
    for file in files:
        # print(root,file)
        file_path = os.path.join(root, file)
        try:
            img = tf.io.decode_image(tf.io.read_file(file_path))
        except:
            # print("corrupt: ", file_path)
            os.remove(file_path)
            corrupts += 1
print(corrupts)
```

```python
train_ds = image_dataset_from_directory(dr, validation_split=0.2, image_size=(128, 128), batch_size=32, subset="training", label_mode='categorical', seed=seed)
```

```python
test_ds = image_dataset_from_directory(dr, validation_split=0.2, image_size=(128, 128), batch_size=32, subset="validation", label_mode='categorical', seed=seed)
```
