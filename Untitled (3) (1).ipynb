```python
classes_name=['t-shirt','trouser','pullover','dress','coat','sandal','shirt','sneaker','bag','ankleboot']
```

```python
# eda=exploratory data analysis
# 2x5 5x2 img layout
plt.figure(figsize=(10,5))
for i in range(10):
    plt.subplot(2,5,i+1)
    # plt.title(f"{train_y[i]}")
    plt.title(f"{classes_name[train_y[i]]}")
    plt.imshow(train_x[i],cmap=plt.cm.binary)
plt.tight_layout()
```

```python
# colour channel same krna 
# resize krna
# normalize krna jo img h divide by 255
```

```python
train_x=train_x/255.0
test_x=test_x/255.0
```

```python
from tensorflow.keras.layers import Dense,Flatten,Conv2D,GlobalMaxPooling2D
# dense or flatten layer aa gyi
from tensorflow.keras import Sequential
# layers se sth krna h vo
from tensorflow.keras.optimizers import SGD
# loss kam krega weight update krke
```

```python
# ann use kr rhe mtlb img pixdel as it is jane chahiye 
# flatten data ko same ek size kr deta h 28x28 ko 784 me kr dega 
model1=Sequential([
    Conv2D(32,(3,3),activation="relu",input_shape=(28,28,1)),
    GlobalMaxPooling2D(),
    Flatten(),
    Dense(128,activation="relu"),   #no. of neuron diye or func diya hidden layer bani
    Dense(10,activation="softmax")  #out layer bani softmax isiliye kyu ki jyda neuron h sigmoid ni kyu ki vo sirf binary leta h value 
])
model1.summary()
# input*hidden+bias isme hoga 784*128+128(parameter kitne h )(weight matrix kya h vo pta chla layer k bich me )
```

```python
model=Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128,activation="relu"),   #no. of neuron diye or func diya hidden layer bani
    Dense(10,activation="softmax")  #out layer bani softmax isiliye kyu ki jyda neuron h sigmoid ni kyu ki vo sirf binary leta h value 
])
model.summary()
```

```python
model2=Sequential([
    Conv2D(32,(3,3),activation="relu",input_shape=(28,28,1)),
    Flatten(),
    Dense(128,activation="relu"),   #no. of neuron diye or func diya hidden layer bani
    Dense(10,activation="softmax")  #out layer bani softmax isiliye kyu ki jyda neuron h sigmoid ni kyu ki vo sirf binary leta h value 
])
model2.summary()
```

```python
from numpy.matrixlib.defmatrix import matrix
```

```python
# optimizer weight update krne k liye
opt=SGD(learning_rate=0.01)
```

```python
# compileing model
model.compile(opt,tf.keras.losses.sparse_categorical_crossentropy,metrics=['accuracy'])
```

```python
# train kr rhe model ko
model.fit(train_x,train_y,epochs=20,validation_split=0.2)
# val_accuracy0.8 se jyda ya vo aani chahiye
```

```python
model.evaluate(test_x,test_y)
# 
```
