```python
# Input & target
x = 0.5
t = 1
lr = 0.1

# Weights (just simple values)
w1, w2 = 0.2, -0.3          # input → hidden1
w3, w4 = 0.4, 0.1           # hidden1 → hidden2
w5, w6 = -0.2, 0.3
w7, w8 = 0.2, -0.5          # hidden2 → output

# Bias
b = 0.1

# Sigmoid
def sig(x):
    return 1 / (1 + 2.71828**(-x))

def dsig(y):
    return y * (1 - y)

# -------- Forward --------
h1 = sig(x*w1 + b)
h2 = sig(x*w2 + b)

h3 = sig(h1*w3 + h2*w5 + b)
h4 = sig(h1*w4 + h2*w6 + b)

out = sig(h3*w7 + h4*w8 + b)

print("Before:", out)

# -------- Error --------
e = t - out

# -------- Backprop --------
d_out = e * dsig(out)

d_h3 = d_out * w7 * dsig(h3)
d_h4 = d_out * w8 * dsig(h4)

d_h1 = (d_h3*w3 + d_h4*w4) * dsig(h1)
d_h2 = (d_h3*w5 + d_h4*w6) * dsig(h2)

# -------- Update --------
w7 += lr * d_out * h3
w8 += lr * d_out * h4

w3 += lr * d_h3 * h1
w5 += lr * d_h3 * h2

w4 += lr * d_h4 * h1
w6 += lr * d_h4 * h2

w1 += lr * d_h1 * x
w2 += lr * d_h2 * x

b += lr * d_out   # simple bias update

# -------- After --------
h1 = sig(x*w1 + b)
h2 = sig(x*w2 + b)

h3 = sig(h1*w3 + h2*w5 + b)
h4 = sig(h1*w4 + h2*w6 + b)

out = sig(h3*w7 + h4*w8 + b)

print("After:", out)
```
