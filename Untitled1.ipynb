```python
import numpy as np
import matplotlib.pyplot as plt

# Input image
image = np.array([
    [1, 2, 3, 0, 1],
    [4, 5, 6, 1, 2],
    [7, 8, 9, 2, 3],
    [1, 3, 5, 4, 0],
    [2, 4, 6, 1, 1]
])

# Kernel
kernel = np.array([
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1]
])
# Function to compute output size
def get_output_size(N, F, P, S):
    """
    N = input size
    F = filter/kernel size
    P = padding
    S = stride
    """
    return ((N - F + 2 * P) // S) + 1


# Function to compute padding when desired output size is known
def get_padding(N, F, S, O):
    """
    N = input size
    F = filter size
    S = stride
    O = desired output size
    """
    return ((O - 1) * S - N + F) // 2


# Function to compute SAME padding (stride=1)
def get_same_padding(F):
    """
    F = filter size
    """
    return (F - 1) // 2
# Same padding
pad = 1
padded = np.pad(image, pad, mode='constant')

# Output
output = np.zeros_like(image)

# Manual convolution
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        region = padded[i:i+3, j:j+3]
        output[i, j] = np.sum(region * kernel)

# Plotting
fig, ax = plt.subplots(1,4, figsize=(14,4))

ax[0].imshow(image, cmap='gray')
ax[0].set_title("Input Image")

ax[1].imshow(kernel, cmap='gray')
ax[1].set_title("3x3 Kernel")

ax[2].imshow(padded, cmap='gray')
ax[2].set_title("Padded Image")

ax[3].imshow(output, cmap='gray')
ax[3].set_title("Output Feature Map")

for a in ax:
    a.axis("off")

plt.tight_layout()
plt.show()
```
