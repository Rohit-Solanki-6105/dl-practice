```python

# -------- Image --------
img = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

kernel = [
    [1,0],
    [0,-1]
]

# -------- 1. Padding --------
def padding(img, pad):
    n = len(img)
    new_n = n + 2*pad

    out = [[0]*new_n for _ in range(new_n)]

    for i in range(n):
        for j in range(n):
            out[i+pad][j+pad] = img[i][j]

    return out


# -------- 2. Convolution --------
def convolution(img, kernel, stride=1):
    n, f = len(img), len(kernel)
    out = []

    for i in range(0, n - f + 1, stride):
        row = []
        for j in range(0, n - f + 1, stride):

            val = sum(
                img[i+ki][j+kj] * kernel[ki][kj]
                for ki in range(f)
                for kj in range(f)
            )

            row.append(val)
        out.append(row)

    return out


# -------- 3. Max Pooling --------
def max_pooling(img, size=2, stride=2):
    n = len(img)
    out = []

    for i in range(0, n - size + 1, stride):
        row = []
        for j in range(0, n - size + 1, stride):

            # take max from window
            window_max = max(
                img[i+ki][j+kj]
                for ki in range(size)
                for kj in range(size)
            )

            row.append(window_max)
        out.append(row)

    return out


# -------- Run --------
padded = padding(img, 1)
print("Padded:")
for r in padded:
    print(r)

conv_out = convolution(padded, kernel, stride=1)
print("\nConvolution:")
for r in conv_out:
    print(r)

pool_out = max_pooling(conv_out, 2, 2)
print("\nPooling:")
for r in pool_out:
    print(r)
```
