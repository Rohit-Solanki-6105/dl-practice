```python
from tensorflow.keras.layers import Dense,Flatten,Conv2D,GlobalMaxPooling2D
# dense or flatten layer aa gyi
from tensorflow.keras import Sequential
# layers se sth krna h vo
from tensorflow.keras.optimizers import SGD
# loss kam krega weight update krke
```

```python
model1=Sequential([
    Conv2D(32,(3,3),activation="relu",input_shape=(28,28,1)),
    GlobalMaxPooling2D(),
    Flatten(),
    Dense(128,activation="relu"),   #no. of neuron diye or func diya hidden layer bani
    Dense(10,activation="softmax")  #out layer bani softmax isiliye kyu ki jyda neuron h sigmoid ni kyu ki vo sirf binary leta h value 
])
model1.summary()
```

```python
opt=SGD(learning_rate=0.01)
```

```python
# compileing model
model1.compile(opt,tf.keras.losses.sparse_categorical_crossentropy,metrics=['accuracy'])
```

```python
# train kr rhe model ko
model1.fit(train_ds,epochs=20,validation_split=0.2)
# val_accuracy0.8 se jyda ya vo aani chahiye
```
