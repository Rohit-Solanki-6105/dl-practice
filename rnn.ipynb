```python
from nltk.corpus import movie_reviews
```

```python
import nltk
```

```python
nltk.download("movie_reviews")
```

```python
for i in movie_reviews.fileids():
    print(i)
```

```python
file = movie_reviews.fileids()[0]
print(movie_reviews.raw(file))
```

```python
documents = [(movie_reviews.raw(fileid), cat) for cat in movie_reviews.categories() for fileid in movie_reviews.fileids(categories=cat)]
```

```python
reviews, labels = [documents[i][0] for i in range(len(documents))], [documents[i][1] for i in range(len(documents))]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
df['reviews'] = reviews
df['labels'] = labels
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
```

```python
data = LE.fit_transform(df['labels'])
```

```python
df['le'] = LE.fit_transform(df['labels'])
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tk = Tokenizer()
```

```python
tk.fit_on_texts(df['reviews'])
```

```python
seq = tk.texts_to_sequences(df['reviews'])
```

```python
print(seq[0])
```

```python
tk.word_index
```

```python
vocab_size = len(tk.word_index)
```

```python
print(len(seq[0]), len(seq[1]))
```

```python
import numpy as np
```

```python
max_length = np.max([len(i) for i in seq])
```

```python
## padded sentence
x = pad_sequences(seq, maxlen=max_length )
```
